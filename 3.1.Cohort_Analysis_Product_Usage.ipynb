### **Cohort Analysis - Usage Frequency**

In this document, we will explore the frequency of usage of Ironhack's students throughout a period of time.
The main objective of performing such analysis is to discover any patterns that could help making business decisions.

Some potential questions to be answered are the following:
- Are students using the service less because they are manually reviewed too much?
- Are there specific months into the journey where students use the service more?
- Are there any cohorts using the service more worth exploring?

```python
#We star by importing the required libraries:

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.colors as mcolors
```

```python
#Importing the different datasets

cash = pd.read_csv('/home/ubuntu/Ironhack_all/Ironhack_2ndProject/project_dataset/cash-clean.csv')
fees = pd.read_csv('/home/ubuntu/Ironhack_all/Ironhack_2ndProject/project_dataset/fees-clean.csv')
```

#### **Performing data checks, cleaning & transformation**

Before starting any analysis, we need to check for data integrity as well as the quality. Appropriate procedures are conducted to fix or repair any missing values.
Afterwards, a subset is created from the original data to be used for later analyses.


```python
#Creating a subset with the necessary columns only
cash_cohort = cash[['id','user_id','created_at','deleted_account_id','recovery_status', 'status']]

#Make sure no more missing user_ids
cash_cohort[cash_cohort['user_id'].isna()]
```

#### **Cohort Analysis**

In this section, we will define the variable/column on which the cohort will be based. In this case, minimum cash request date is taken as the cohort group. The analysis will essentially be looking at the people that created a cash request, how much % of the same people keep using repeatedly the service.


```python
#I'm transforming the "created_at" to datetime for manipulation afterwards
cash_cohort['created_at'] = pd.to_datetime(cash_cohort['created_at'])


cash_cohort['cohort group'] = cash_cohort.groupby('user_id')['created_at'].transform('min').dt.to_period('M') #Getting the minimum date from each unique person, and getting YYYY-MM
cash_cohort['usage date'] = cash_cohort['created_at'].dt.to_period('M')  #Getting the dates they make cash request in YYYY-MM too (as they can make as many as they want)

#Calculating how many months have passed between creation date and another cash request. If they're the same, it will be 0
cash_cohort['Passed months'] = (cash_cohort['usage date'].dt.year - cash_cohort['cohort group'].dt.year)*12 + cash_cohort['usage date'].dt.month - cash_cohort['cohort group'].dt.month

cash_cohort
```

```python
cash_cohort[cash_cohort['user_id']==13681.0] #Checking a random user to see that in fact this user has more than 1 CR
```

```python
groupped_cohort = cash_cohort.groupby(['cohort group', 'Passed months'])['user_id'].nunique().reset_index()

#I'm groupping by cohort, and passed months and then I'm trying to see how many unique users have made cash request.

cohort_pivot = groupped_cohort.pivot(index='cohort group', columns='Passed months', values='user_id') #Pivoting the group by

Initial_users = cohort_pivot[0] 
#I'm extracting this column because I want to know the number total students by cohort group. 
#So later we can compare the % of students from the total that made another cash request
Initial_users
```

```python
cohort_graph = cohort_pivot.divide(cohort_pivot.iloc[:, 0], axis=0) # Storing a version for graph later (it will deal with NAns automatically and put %)
cohort_pivot_n = cohort_pivot.divide(cohort_pivot.iloc[:, 0], axis=0) * 100 #This is for the table, so I have to multiply by 100 to get % scale
```

```python
cohort_pivot_n.iloc[0] = cohort_pivot.iloc[0].fillna(0) #filling specific spots with 0 to differentiate actual 0s to no data yet  
cohort_pivot_n.loc['2020-02',9] = 0
cohort_pivot_n.loc['2020-04',7] = 0

cohort_graph.iloc[0] = cohort_pivot.iloc[0].fillna(0) #filling specific spots with 0 to differentiate actual 0s to no data yet  
cohort_graph.loc['2020-02',9] = 0
cohort_graph.loc['2020-04',7] = 0

cohort_pivot_n
```

```python
cohort_graph
```

```python
cohort_pivot_n = cohort_pivot_n.applymap(lambda x: '{:.2f}%'.format(x) if pd.notna(x) else '' ) #Applying a lambda function to get every single number in % format

cohort_pivot_n
```

```python
cohort_pivot_n.insert(0,'Total_users', Initial_users) #As mentioned in line[9], I'm inserting total users to see the number of students by cohort-group
cohort_pivot_n
```

```python
with sns.axes_style("white"):
    fig, ax = plt.subplots(1, 2, figsize=(12, 8), sharey=True, gridspec_kw={'width_ratios': [1, 11]})
    
    # cohort_graph heatmap itself
    sns.heatmap(cohort_graph, 
                mask=cohort_graph.isnull(), 
                annot=True, 
                fmt='.0%', 
                cmap='RdYlGn', 
                ax=ax[1])
    ax[1].set_title('Monthly Cohorts: Usage Frequency', fontsize=16)
    ax[1].set(xlabel='# of periods',
              ylabel='')

    # This is where I include the total number of users so we have it in the graph as well
    cohort_size_df = pd.DataFrame(Initial_users).rename(columns={0: 'Months passed'})
    white_cmap = mcolors.ListedColormap(['white'])
    sns.heatmap(cohort_size_df, 
                annot=True, 
                cbar=False, 
                fmt='g', 
                cmap=white_cmap, 
                ax=ax[0])

    fig.tight_layout()
```

#### **Cohort Analysis - Interpretation**

1) The first cohort(2019-11) doesn't provide much valuable info as the whole cohort is just one person, so we will disregard this cohort group.

2) Cohort '2020-05' has the best % of usage frequency after its creation month.

3) We can see that the first two cohorts have not so great retention/frequency usage rate as the % is always in the 20s after the first created month. In comparison, we see an improvement in the cohort '2020-02' and onwards. Therefore, a quick analysis is done to check whether manual review of cash request(indicated by the rows where moderated_at is not null) impacts the % of retention. A comparison is done between cohorts from '2019-12' to '2020-05'


```python
deep_cohort = cash[['id','user_id','created_at','deleted_account_id','recovery_status','moderated_at', 'status', 'transfer_type']] #selecting again the specific columns needed
deep_cohort['moderated_at'] = deep_cohort['moderated_at'].isna() #Changing the column to boolen "True" or "False" (check if it is blank) as blank means no manual review

#The following code is used to create the cohort again
deep_cohort['created_at'] = pd.to_datetime(deep_cohort['created_at'])
deep_cohort['usage date'] = deep_cohort['created_at'].dt.to_period('M')

deep_cohort['cohort'] = deep_cohort.groupby('user_id')['created_at'].transform('min').dt.to_period('M')

deep_cohort

```

```python

#This code is to select only the cohorts I want to compare. Only selecting these cohorts because there's not so much data after '2020-06'
deep_cohort = deep_cohort[(deep_cohort['cohort'] == '2019-12') |
                          (deep_cohort['cohort'] == '2020-01') | 
                          (deep_cohort['cohort'] == '2020-02') | 
                          (deep_cohort['cohort'] == '2020-03') | 
                          (deep_cohort['cohort'] == '2020-04') | 
                          (deep_cohort['cohort'] == '2020-05') | 
                          (deep_cohort['cohort'] == '2020-06')]

#Code 1 to create a groupped subset to analyze differences between manual reviewed requests and those not reviewed
deep_manual = deep_cohort.groupby(['cohort','moderated_at'])['user_id'].count().reset_index() 

#Code 2 to create a groupped subset to analyze differences between different status
deep_status = deep_cohort.groupby(['cohort','status'])['user_id'].nunique().reset_index()

#Code 3 to create a groupped subset to analyze differences between different transfer types
deep_transfer = deep_cohort.groupby(['cohort','transfer_type'])['user_id'].count().reset_index()

deep_status
```

```python
deep_manual['total_count'] = deep_manual.groupby(['cohort'])['user_id'].transform('sum') #counting the sum of False + True

deep_manual['percentage'] = deep_manual['user_id']/deep_manual['total_count']*100 #calculating the percentage over the total

deep_group = deep_manual.rename(columns={'user_id': 'count of manual reviews', 'total_count' : 'total'}) #renaming some columns

deep_group
```

```python
deep_status['total_count'] = deep_status.groupby(['cohort'])['user_id'].transform('sum') #calculating the sum of all status

deep_status['percentage'] = deep_status['user_id']/deep_status['total_count']*100 #calculating the percentage

deep_status = deep_status[deep_status['status'].isin(['rejected', 'money_back'])] #I only want to see money-back and rejected status

deep_group2_pivot = deep_status.pivot(index='cohort', columns = 'status', values = 'percentage') #pivoting

deep_group2_pivot
```

```python
#Plotting a line graph to clearly visualize patterns

plt.figure(figsize=(10,6))

deep_group2_pivot.plot(kind='line', marker='o', ax=plt.gca())

plt.title('Percentage of Cash Requests per Status and Cohort')
plt.xlabel('Cohort')
plt.ylabel('Percentage per status')
plt.xticks(rotation=0)
plt.legend(title = 'Status of CR')

plt.show()

```

```python
deep_transfer
```

```python
deep_transfer['total_count'] = deep_transfer.groupby(['cohort'])['user_id'].transform('sum') #summing all transfer types, calculating total

deep_transfer['percentage'] = deep_transfer['user_id']/deep_transfer['total_count']*100 #calculate % of instant & regular

deep_transfer_pivot = deep_transfer.pivot(index='cohort', columns='transfer_type', values='percentage') #pivoting

deep_transfer_pivot
```

```python
#Plotting bar graph to see trends

plt.figure(figsize=(10,6))

deep_transfer_pivot.plot(kind= 'bar', ax=plt.gca())

plt.title('Percentage of Cash Requests per Transfer type and Cohort')
plt.xlabel('Cohort')
plt.ylabel('Percentage per status')
plt.xticks(rotation=0)
plt.legend(title = 'Transfer type')

plt.show()
```

```python
#Based on previous hypotheses, we want to assess whether the rejection rate within the first month of cash request impacts later usage.
#Here we want to group by cash creation date

deep_rejection = deep_cohort[(deep_cohort['usage date'] == '2019-12') |
                          (deep_cohort['usage date'] == '2020-01') | 
                          (deep_cohort['usage date'] == '2020-02') | 
                          (deep_cohort['usage date'] == '2020-03') | 
                          (deep_cohort['usage date'] == '2020-04') | 
                          (deep_cohort['usage date'] == '2020-05') | 
                          (deep_cohort['usage date'] == '2020-06')]

groupped_rejection = deep_rejection.groupby(['usage date','status'])['user_id'].nunique().reset_index()

groupped_rejection['status'] = groupped_rejection['status'].apply(lambda x: 'not rejected' if x != 'rejected' else x)

groupped_rejection['total_count'] = groupped_rejection.groupby(['usage date'])['user_id'].transform('sum')

groupped_rejection['rejection_rate'] = groupped_rejection['user_id'] / groupped_rejection['total_count'] * 100

groupped_rejection = groupped_rejection[groupped_rejection['status'] == 'rejected']

groupped_rejection
```

```python
groupped_rejection_pivot = groupped_rejection.pivot(index='usage date', columns='status', values='rejection_rate')

plt.figure(figsize=(10,6))

groupped_rejection_pivot.plot(kind='line', marker='o', ax=plt.gca())

plt.title('Rejection rate in first month')
plt.xlabel('Cohort')
plt.ylabel('Percentage of rejection')
plt.xticks(rotation=0)

plt.show()
```

From our analysis, we can see:
1) That the drop in usage frequency can't really be atributted to whether the request has been reviewed manually or not. The initial hypothesis was that maybe if a cohort group has had many manual reviews, then the usage frequency will drop. From the numbers it doesn't seem to matter as the % of manual reviews between a cohort that has great retention and a cohort that doesn't, does not vary too much

2) We see that cohort groups where retention is great (e.g: '2020-02' or '2020-03') have more people within the status "money back" and fewer people in "rejected". This could suggest that if the journey is successful (no rejects, and student successfully pays back), then the satisfaction is great prompting them to come back and use the service in the future.

3) We see that request for regular payments have been decreasing over cohort-groups, and coincidentally, the retention rate seems to be higher. This could potentially indicate an area to focus on.

4) To test the hypothesis in section '2', we have run some more in-depth analysis in line 84 and we see that indeed, the lower the rejection rate, the better the retention gets. The best performing month in terms of product usage is the cohort '2020-05', we see that the first month, the retention rate was 59%! (this coincides with the minimum low of rejection rate).
